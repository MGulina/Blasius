## Avant-propos

Comprendre la Nature… Ces trois mots incarnent la pensée du Physicien. Dans ce but, il commence par étudier les solides, autrement plus simples à décrire que les fluides, ils ont tout de même beaucoup à nous dire. En Dynamique des Fluides, on étudiera des objets de nature bien plus complexe : tantôt calmes et tranquilles, tantôt impétueux et tumultueux, les fluides, quant à eux, ont encore plus de choses à nous dire ! La simplicité du solide n’est plus ; ce sont des équations aux dérivées partielles non linéaires qui régissent le mouvement des fluides « newtoniens » comme milieux continus. Elles furent écrites par Navier et Stokes dans le courant du $XIX^{ème}$ siècle et prennent, si l’on suppose l’incompressibilité, la forme suivante :

$$
   \rho \left[ \frac{\partial \vec{u}}{\partial t}  + \left( \vec{u} \cdot \vec{\nabla} \right) \vec{u} \right] =  - \vec{\nabla}P + \mu \Delta \vec{u} + \vec{f}_{ext}
$$

De nombreux artistes de la Science ont contribué à la résolution de ces équations dans différents cas particuliers. Nous étudions ici l'un de ces cas : le profil de vitesse sur une plaque plane. Partant des équations de Navier-Stokes, Prandtl trouva des équations aux dérivées partielles. C'est alors que Blasius, étudiant de Prandtl, proposa une formulation en une équation différentielle ordinaire ...

### 1.1 - Équations de Prandtl - Blasius

Considérons une plaque plane et modélisons-la par le demi-plan infini généré par les axes orthogonaux habituels $x$ et $z$ de sorte qu'il soit dans la partie positive de $x$. Le $3^{ème}$ axe, $y$, pointe donc vers le haut, perpendiculairement à ce plan.

Ajoutons maintenant les vents, soufflant à la vitesse $\vec{U}$ dans la direction $x$. Une fois arrivé dans la partie positive de l'axe, la vitesse du fluide subira l'influence de la plaque et on la notera $\vec{u} = \left( u_x(t,x,y),u_y(t,x,y) \right)$ où l'on peut se passer de la dépendance en $z$ par symétrie de translation. Notons que pour la suite, nous ferons toujours l'hyptothèse d'être à grand nombre de Reynolds.

#### 1.1.1 - Formulation de Prandtl (1875 - 1953) en équations aux dérivées partielles

##### Analyse de l'équation pour $u_x$

Écrivons l'équation de Navier-Stokes sans forces externes projetée sur $x$ en divisant les deux membres par $\rho$ pour avoir, sachant que la viscosité dynamique est définie comme $\nu = \mu/\rho$ : 

$$
  \frac{\partial u_x}{\partial t}  + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} =  
  -  \frac{1}{\rho}\frac{\partial P}{\partial x} + \nu \frac{\partial^2 u_x }{\partial x^2} + \nu \frac{\partial^2 u_x }{\partial y^2}
$$

On peut supposer, à grand nombre de Reynolds, que :

$$ \left\{ \;    
\begin{array}{lcl}
\left| \partial_x u_x \right|    & << & \left| \partial_y u_x \right| \\ 
\left| \partial_x^2 u_x \right|  & << & \left| \partial_y^2 u_x \right| \\ 
\end{array} 
\right.
$$

Ces inégalités rendent licite l'écriture suivante pour la composante $u_x$ :

$$
  \frac{\partial u_x}{\partial t}  + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} =
  - \frac{1}{\rho}\frac{\partial P}{\partial x} + \nu \frac{\partial^2 u_x }{\partial y^2} 
$$

Où l'on a pu négliger l'un des termes du laplacien. Il est important de remarquer que l'on n'a pas simplifié les dérivées premières malgré la tentation. En effet, en regardant attentivement on s'aperçoit que la pondération résultante de l'opérateur $\vec{u} \cdot \vec{\nabla}$ confère à ces deux termes un même ordre de grandeur. Voyons cela, l'incompressibilité $(\partial_t \rho = 0)$ force l'écoulement à être solénoïdal à travers l'équation de continuité :

$$
\vec{\nabla} \cdot \vec{u} = 0
$$

Si on pose $L$ comme la distance caractéristique dans la direction $x$ et $\delta$  l'analogue pour la direction $y$, on peut écrire cette contrainte de manière dimensionnelle :

$$
\frac{u_x}{L} \sim \frac{u_y}{\delta}
$$

Il s'en suit les égalités ci-dessous :

$$ \left\{ \;    
\begin{array}{lclcl}
\left| u_x \partial_x u_x \right| & \sim & U^2/L       &      &         \\ 
\left| u_y \partial_y u_x \right| & \sim & u_yu_x/\delta & \sim & U^2/L \\ 
\end{array} 
\right.
$$

##### Expression de $\delta$

À la limite où l'on a égalité des contributions du terme non linéaire et du terme visqueux, il vient :

$$
 \frac{U^2}{L} \sim \nu \frac{U}{\delta^2} \Longrightarrow \fbox{$\delta \sim \frac{L}{\sqrt{Re}}$}
$$

En utilisant la définition du nombre de Reynolds $Re = UL/\nu$. On appelle <i>couche limite</i> la zone en contact avec la plaque et d'épaisseur $\delta$. C'est dans cette couche que les effets visqueux ne peuvent pas être négligés.

##### Analyse de l'équation pour $u_y$

Exploitons maintenant l'équation pour $u_y$. Celle-ci s'écrit comme :

$$
  \frac{\partial u_y}{\partial t}  + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} =
  - \frac{1}{\rho}\frac{\partial P}{\partial y} + \nu \frac{\partial^2 u_y }{\partial x^2} + \nu \frac{\partial^2 u_y }{\partial y^2} 
$$

L'analyse dimensionnelle fournit les relations suivantes, avec $T \sim L/u_x$ :

$$ \left\{ \;    
\begin{array}{lclcl}
\left| \partial_t u_y \right|       & \sim & u_x\delta/LT                            & \sim & \delta U^2/L^2    \\ 
\left| u_x \partial_x u_y \right|   & \sim & u_x u_y/L                               & \sim & \delta U^2/L^2    \\
\left| u_y \partial_y u_y \right|   & \sim & u_y^2/\delta                            & \sim & \delta U^2/L^2    \\
\left| \nu \partial_x^2 u_y \right| & \sim & \frac{\nu}{L^2}     \frac{u_x\delta}{L} & \sim & \delta^3 U^2/L^4  \\
\left| \nu \partial_y^2 u_y \right| & \sim & \frac{\nu}{\delta^2}\frac{u_x\delta}{L} & \sim & \delta U^2/L^2    \\
\end{array} 
\right.
$$

Par compatibilité avec l'équation pour $u_x$, on ne conserve que les termes en $\mathcal{O}(\delta) = 0$. De cette équation on tire finalement que $p = p(x)$ puisque l'on a $\left| \partial_y P \right| \sim \rho \delta U^2/L^2$. 

##### Équation pour la pression

On prend alors pour $p(x)$, la pression sur le bord de la couche limite. On l'obtient en solutionnant l'équation de Navier-Stokes dans la zone libre de viscosité : 

$$
  \frac{\partial u_x}{\partial t}  + u_x\frac{\partial u_x}{\partial x} = - \frac{1}{\rho}\frac{dP}{dx}
$$


##### Hypothèse de stationnarité

Supposons maintenant que $\partial_t u_x = 0 = \partial_t u_y$. L'équation pour la pression se réduit comme : 

$$
  -\frac{1}{\rho}\frac{dP}{dx} = u_x\frac{\partial u_x}{\partial x} = 0
$$

Car loin de la plaque $u_x \to U \Longrightarrow \partial_x U = 0$. Ceci simplifie l'équation de Prandtl restante à l'ordre $0$ en $\delta$ :

$$
u_x \frac{\partial u_x}{\partial x} + u_y \frac{\partial u_x}{\partial y} = \nu \frac{\partial^2 u_x}{\partial y^2}  
$$

##### Équations de Prandtl

Finalement, le système complet des équations de Prandtl à grand nombre de Reynolds ($\mathcal{O}(\delta) = 0$) est obtenu en couplant l'équation ci-dessus à l'équation de continuité sous l'hypothèse d'incompressibilité :

<p style="border:2px; border-style:solid; border-color:blue; padding: 1em; margin-left:36%; margin-right:36%">
$$ \left\{ \;    
\begin{array}{l} 
\vec{\nabla} \cdot \vec{u} = 0   \\
u_x \frac{\partial u_x}{\partial x} + u_y \frac{\partial u_x}{\partial y} = \nu \frac{\partial^2 u_x}{\partial y^2}  \\
\end{array} 
\right.
$$
</p>
    
avec les conditions limites suivantes :
<table style="border:none" width=100% >
    <tr style="border:none">
        <td style="border:none"><div align=right>Non glissement $ $</div></td>
        <td style="border:none">$ $ $u_x(x,0) = 0$</td>
    </tr>

    <tr style="border:none">
        <td style="border:none"><div align=right>Flux entrant uniforme $ $</div></td>
        <td style="border:none">$ $ $u_x(0,y) = U$ et $u_y(0,y) = 0$</td>
    </tr>

    <tr style="border:none">
        <td style="border:none"><div align=right>Imperméabilité de la plaque $ $</div></td>
        <td style="border:none">$ $ $u_y(x,0) = 0$</td>
    </tr>
    
    <tr style="border:none">
        <td style="border:none"><div align=right>Vitesse uniforme à l'infini $ $</div></td>
        <td style="border:none">$ $ $u_x(x,y \to \infty) \to U$</td>
    </tr>    
</table>

#### 1.1.2 - Formulation de Blasius (1883 - 1970) en équation différentielle ordinaire

##### Fonction de courant

Utilisons le formalisme de la fonction de courant que l'on note $\psi(x,y)$ : 

$$ \left\{ \;    
\begin{array}{l} 
u_x =  \partial_y \psi \\
u_y = - \partial_x \psi  \\
\end{array} 
\right.
$$

Injectons ceci dans l'équation de Prandtl pour obtenir : 

$$
\frac{\partial \psi}{\partial y} \frac{\partial^2 \psi}{\partial x \partial y}
- \frac{\partial \psi}{\partial x} \frac{\partial^2 \psi}{\partial y^2}
= \nu \frac{\partial^3 \psi}{\partial y^3}
$$

##### Équation de Blasius

Exprimons maintenant le génie de Blasius en posant la forme de la fonction de courant comme :

$$
 \psi = \sqrt{\nu U x} \; f(\eta) \; ; \; \eta(x,y) = \sqrt{\frac{U}{\nu}}\frac{y}{\sqrt{x}} 
$$

En effet, il avait remarqué que l'épaisseur de la couche limite devait augmenter de manière moins que linéaire et que le profil se modifiait en dilatant continûment sa courbure. Examinons la nouvelle forme des dérivées partielles :

$$
\left\{ \;    
\begin{array}{lllcl}
\partial_x  =  \frac{\partial \eta}{\partial x} \partial_\eta
 = - \frac{1}{2} \frac{\eta}{x} \partial_\eta
 & \Longrightarrow u_y = & - \partial_x \psi 
 & = & \frac{1}{2} \sqrt{\frac{\nu U}{x}}\left( \eta f' - f \right)\\
\partial_y  =  \frac{\partial \eta}{\partial y} \partial_\eta
 = \sqrt{\frac{U}{\nu x}} \partial_\eta
 & \Longrightarrow u_x = & \partial_y \psi & = & U f'\\
\end{array}
\right.
$$

L'équation pour $\psi$ se transforme en :

$$
U f' \left( - \frac{U}{2 x} \eta f'' \right)
+ \frac{1}{2}\sqrt{\frac{\nu U}{x}}\left( \eta f' - f \right) U f'' \sqrt{\frac{U}{\nu x}}
= \nu \frac{U^2}{\nu x} f'''
$$

Après simplifications, on a l'équation de Blasius :

<p style="border:2px; border-style:solid; border-color:blue; padding: 1em; margin-left:40%; margin-right:40%">
$$
   f^{'''} + \frac{1}{2}ff^{''} = 0
$$
</p>

avec les conditions limites suivantes : 
 
$$ \left\{ \;    
\begin{array}{lcl}
f(0) & = & 0 \\ 
f'(0) & = & 0 \\ 
f''(0) & = & ? \; \Longrightarrow \; f'(\infty) & = & 1 \\ 
\end{array}
\right.$$

Car on a supposé le non-glissement et le caractère uniforme de la vitesse loin de la plaque.

##### Comportement de l'écoulement près de la plaque

Pour les petites valeurs de $\eta$, on peut développer $f'$ en série de Taylor autour de $0$ comme : 

$$
f'(\eta) = f'(0) + \eta f''(0) + \mathcal{O}(\eta^2)
$$

Compte tenu des conditions initiales cette expression se réduit à : 

$$
f'(\eta) = \eta f''(0) + \mathcal{O}(\eta^2) \Longrightarrow \fbox{$u_x = U \eta f''(0) + \mathcal{O}(\eta^2)$}
$$

Nous aurons l'occasion de vérifier ce comportement plus loin.

### 1.2 - Méthodes de résolution numérique

Il arrive souvent en physique que des équations ne peuvent être solutionnées analytiquement. Les fameuses équations de Navier-Stokes en donnent un bel exemple. Le problème que nous avons à résoudre en découle et peut s’écrire sous deux formes analytiques équivalentes. Celles-ci sont propices à l’utilisation de méthodes numériques différentes (les différences finies et Runge-Kutta 4) de sorte que nous puissions, à la fin, les comparer.

#### 1.2.1 - Différences finies

La méthode des différences finies (voir <a href="#ref4" style="text-decoration:none">[4]</a> et <a href="#ref6" style="text-decoration:none">[6]</a>) est très intuitive. Elle est basée sur la définition de la dérivée d'une fonction (que l'on suppose, bien entendu, suffisamment lisse) : 

$$
    f'(x) = \lim_{\Delta x \, \to \, 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$

On construit une grille de pas $h$ et on approxime ainsi la dérivée de la fonction à l'aide du nœud de la grille suivant : 

$$
    f'(x) = \frac{f(x + h) - f(x)}{h} + \mathcal{O}(h)
$$

C'est la formule d'Euler progressive. Si on exploite le nœud précédent, on obtient la formule d'Euler rétrograde : 

$$
    f'(x) = \frac{f(x) - f(x - h)}{h} + \mathcal{O}(h)
$$

Afin de gagner en précision, on peut utiliser les deux informations en les moyennant pour avoir la différence centrée : 

$$
    f'(x) = \frac{f(x + h) - f(x - h)}{2h} + \mathcal{O}(h^2)
$$

L'article de B. Fornberg <a href="#ref3" style="text-decoration:none">[5]</a> fournit les coefficients à utiliser pour les dérivées d'ordre supérieur (jusque 4) ainsi que pour des niveaux plus fin de précision.

<a href="#Contenu" style="text-decoration:none">↑ Contenu ↑</a>
</div>

#### 1.2.2 - Intégration numérique d'equations différentielles ordinaires

Considérons le problème de Cauchy suivant : 

$$ \left\{ \;
\begin{array}{lcl}
y'(x) & = & f(x,y(x)) \\ 
y(x_0) & = & y_0 \\ 
\end{array} 
\right.$$

De nouveau, on suppose la fonction $y$ suffisamment lisse sur l'intervalle $I$ que parcours $x$. Dès lors, on peut montrer l'existence et l'unicité d'une solution locale. Celle-ci s'écrira formellement comme : 

$$
y(x) = y_0 + \int\limits_{x_0}^{x} f(x',y(x')) dx'
$$

Maintenant, on discrétise l'intervalle d'intégration en $N + 1$ nœuds et on note le pas $h$ : 

$$
x_n = x_0 + nh  \; ; \; n \in \left\{ 0, 1, \cdots, N \right\}
$$

On définit alors pour chacun de ces nœuds :

$$
y_n = y(x_n) \; ; \; f_n = f(x_n, y_n)
$$

Ceci permet de construire un schéma d'intégration générique calculant successivement $y_n$ en chaque nœud depuis la condition initiale connue : 

$$
y_{n+1} = y_n + hF(x_i, y_i, f_i; h)
$$

Où $F$ est la fonction d'incrément caractérisant le schéma d'intégration utilisé.

Notons que l'intégration numérique est assujettie aux conditions initiales. Pour l'équation de Blasius qui est du $3^{ème}$ ordre, nous n'en avons que 2 sur 3. Cependant, on a tout de même une condition supplémentaire qui peut être exploitée par la méthode du tir afin de pouvoir intégrer correctement.

<div align="right">
<a href="#Contenu" style="text-decoration:none">↑ Contenu ↑</a>
</div>

#### 1.2.3 - Méthode du tir

Si l'on a une condition limite qui n'est pas fixée à l'origine, en vertu du paragraphe précédent, on voit que l'intégration numérique ne peut pas s'enclencher. Par exemple, considérons un système de trois équations différentielles ordinaires du $1^{er}$ ordre pour $f_1$, $f_2$ et $f_3$ tel que l'on ait :

$$ \left\{ \;
\begin{array}{lcl}
f_1(t_0) & = & f_{10} \\ 
f_2(t_0) & = & f_{20} \\ 
f_3(\tau) & = & f_{3\tau} \\ 
\end{array} 
\right.$$
    
On construit alors un algorithme qui va essayer différentes conditions initiales pour $f_3$ jusqu'à en trouver une telle que l'on ait $f_3(\tau) = f_{3\tau}$.  Cette idée d'essayer différentes conditions initiales fait penser à des tirs successifs, d'où le nom de méthode du tir.

Afin de trouver la bonne condition initiale, on interpole linéairement les deux dernières valeurs de $f_3(\tau)$ afin de trouver une approximation de la condition initiale fournissant $f_3(\tau) \approx f_{3\tau}$ qui permettra l'évaluation de l'itération suivante jusqu'à obtenir la précision voulue sur la condition limite.

### 2.1 - Intégration numérique du problème de Blasius par la méthode de Runge-Kutta 4

#### 2.1.1 - Transformation de l'équation de Blasius en un système d'EDO d'ordre 1
On commence par transformer l'équation de Blasius, qui est une équation différentielle ordinaire du $3^{ème}$ ordre, en un système de trois équations différentielles ordinaires du $1^{er}$ ordre :

$$ \left\{ \;
\begin{array}{lclc}
g_1 & = & f   &                 \\ 
g_2 & = & f'  & \Longrightarrow \\ 
g_3 & = & f'' &                 \\ 
\end{array}
\right.
\left\{ \;
\begin{array}{lcl}
g_1' & = & g_2 \\ 
g_2' & = & g_3 \\ 
g_3' & = & -\frac{1}{2} g_1 g_3 \\ 
\end{array}
\right.
$$

Où l'on a explicitement utilisé l'équation de Blasius pour $g_3'$. On exprime les conditions limites en termes des $g_1$, $g_2$ et $g_3$ :


$$ \left\{ \;    
\begin{array}{lcl}
g_1(0) & = & 0 \\ 
g_2(0) & = & 0 \\ 
\lim\limits_{\eta \to + \infty} g_2(\eta) & = & 1 \\ 
\end{array} 
\right.$$


Il faut modifier la $3^{ème}$condition afin de l'exprimer comme une condition initiale de $g_3$ tout en ayant $\lim\limits_{\eta \to + \infty} g_2(\eta) = 1$.

<div align="right">
<a href="#Contenu" style="text-decoration:none">↑ Contenu ↑</a>
</div>

#### 2.1.2 - Fonction d'incrément pour la méthode Runge-Kutta 4

On explicite maintenant la fonction d'incrément pour la méthode RK4 comme le montre <a href="#ref4" style="text-decoration:none">[4]</a> :

$$
y_{n+1} = y_n + h\left( \frac{K_1 + 2K_2 + 2 K_3 + K_4}{6}\right)
$$

avec 

$$ \left\{ \;    
\begin{array}{lcl}
K_1 & = & f_n \\
K_2 & = & f(x_n + \frac{h}{2}, y_n + \frac{h}{2}K_1) \\
K_3 & = & f(x_n + \frac{h}{2}, y_n + \frac{h}{2}K_3) \\
K_4 & = & f(x_{n+1}, y_n + hK_3) \\
\end{array} 
\right.$$

Notons que le système d'équations à résoudre ici ne fait pas intervenir $x$ explicitement. Cette remarque permet d'alléger légèrement l'implémentation de la méthode. 

<div align="right">
<a href="#Contenu" style="text-decoration:none">↑ Contenu ↑</a>
</div>

#### 2.1.3 - Intégration du système de Blasius

Afin de simplifier la comparaison avec l'algorithme des différences finies, nous avons laissé le choix entre la construction explicite de $\eta$ à partir des variables $x$ et $y$ ou bien la construction directe de $\eta$. Ces deux voies de construction correspondent à build = 0 ou build = 1 (par défaut) respectivement dans le code ci-dessous.

Les valeurs implémentées par défaut permettent à la fois de comparer les graphes obtenus par les deux méthodes et de confronter la valeur $g_3(0)$ à la valeur de référence $0,332057$.

```python
# Librairies
import time  
import numpy as np
from math import sqrt, log, floor
import matplotlib.pyplot as plt
%matplotlib inline

#################################################################

# ----------------------------------------------------------------
# 1 - Intégrateur Runge-Kutta d'ordre 4 pour le système de Blasius
# ----------------------------------------------------------------

# Arguments :
# -----------

    # m               : Nombre de points d'intégration
    # x_min et x_max  : Intervalle d'intégration
    # CIg3            : Condition initiale pour g_3(eta) = f'''(eta)

# Plan
# ----

# 1.1 - Initialisation
    # 1.1.1 - Système d'équations
    # 1.1.2 - Pas d'intégration
    # 1.1.3 - Déclaration des points d'intégration
    # 1.1.4 - Exploitation de la condition initiale non nulle

# 1.2 - Intégration par RK4
    # 1.2.1 - Stockage provisoire des g
    # 1.2.2 - Calcul des termes du schéma RK4 : k = hK 
    # 1.2.3 - Addition des termes avec la pondération du schéma RK4

# ----------------------------------------------------------------    
    
def RK4_Blasius(m, x_min, x_max, CIg3) :
    # 1.1 - Initialisation    
    # 1.1.1 - Système d'équations
    eqn1 = 'Y2'
    eqn2 = 'Y3'
    eqn3 = '-0.5*Y1*Y3'
    
    # 1.1.2 - Pas d'intégration
    h = (x_max - x_min)/(m-1)
    
    # 1.1.3 - Déclaration des points d'intégration
    g1 = np.zeros(m)
    g2 = np.zeros(m)
    g3 = np.zeros(m)

    # 1.1.4 - Exploitation de la condition initiale non nulle
    g3[0] = CIg3
        
    # 1.2 - Intégration par RK4   
    for i in range(0,m-1) : # i va de 0 à m-2

        # 1.2.1 - Stockage provisoire des g
        YY1 = g1[i]
        YY2 = g2[i]
        YY3 = g3[i]

        # 1.2.2 - Calcul des termes du schéma RK4 : k = hK       
        # Terme K1
        Y1 = YY1
        Y2 = YY2
        Y3 = YY3
        
        k11 = h*eval(eqn1)
        k12 = h*eval(eqn2)
        k13 = h*eval(eqn3)
        
        # Terme K2        
        Y1 = YY1 + (k11/2)
        Y2 = YY2 + (k12/2)
        Y3 = YY3 + (k13/2)
        
        k21 = h*eval(eqn1)   
        k22 = h*eval(eqn2)  
        k23 = h*eval(eqn3)
        
        # Terme K3
        Y1 = YY1 + (k21/2)
        Y2 = YY2 + (k22/2)
        Y3 = YY3 + (k23/2)
        
        k31 = h*eval(eqn1)
        k32 = h*eval(eqn2)
        k33 = h*eval(eqn3)
        
        # Terme K4
        Y1 = YY1 + k31
        Y2 = YY2 + k32
        Y3 = YY3 + k33
        
        k41 = h*eval(eqn1)  
        k42 = h*eval(eqn2)
        k43 = h*eval(eqn3)

        # 1.2.3 - Addition des termes avec la pondération du schéma RK4
        g1[i+1] = g1[i] + (k11 + 2*k21 + 2*k31 + k41)/6.
        g2[i+1] = g2[i] + (k12 + 2*k22 + 2*k32 + k42)/6.
        g3[i+1] = g3[i] + (k13 + 2*k23 + 2*k33 + k43)/6.

    return np.transpose(np.array([g1, g2, g3]))

#############################################################

# -----------------------------------------------------------
# 2 - Intégration du système de Blasius par la méthode du tir
# -----------------------------------------------------------

# Plan
# ----

# 2.1 - Paramètrisation
    # 2.1.1 - Viscosité dynamique et vitesse à l'infini
    # 2.1.2 - Définition des vecteurs y et eta
    # 2.1.3 - Tolérance sur la valeur de g_2 en etamax
    
# 2.2 - Intégration couplée au tir (g_1 = g[:,0], g_2 = g[:,1] et g_3 = g[:,2])
    # 2.2.1 - Initialisation des conditions initiales
    # 2.2.2 - Premiers tirs
    # 2.2.3 - Boucle principale

# 2.3 - Approximation de g_2 pour les petites valeurs de eta (0 < eta <~ 4.)
    
# 2.4 - Sorties
    # 2.4.1 - Numériques    
    # 2.4.3 - Graphiques
    
# -----------------------------------------------------------    
    
# 2.1 - Paramètrisation
# 2.1.1 - Viscosité dynamique et vitesse à l'infini
nu = 10**(-5)                 # Modifiable (par défaut 10**(-5) m²/s)
U = 10                        # Modifiable (par défaut 10 m/s)

# 2.1.2 - Construction du vecteur eta
build = 0                     # Modifiable : 0 ou 1 (par défaut 0)
   
xmax = 0.7                    # Modifiable (par défaut 0.7 m)
ymax = 0.006                  # Modifiable (par défaut 0.006 m)

m = 60 + 1                    # Modifiable (par défaut 60+1 pour un pas identique aux D.F.)

y = np.linspace(0, ymax, m)
    
if build == 0 : # Construction de eta via x et y     
    eta_rk4 = sqrt(U/nu)*y/sqrt(xmax/2)
    etamax = eta_rk4[-1]

elif build == 1 : # Construction directe de eta
    etamax = 10               # Modifiable (par défaut 10)
    m = 1000 + 1              # Modifiable (par défaut 1000 + 1)

    eta_rk4 = np.linspace(0, etamax, m)

# 2.1.3 - Tolérance sur la valeur de g_2 en etamax
tol = 10**(-11)               # Modifiable (par défaut 10**(-11))
      
# 2.2 - Intégration couplée au tir (g_1 = g[:,0], g_2 = g[:,1] et g_3 = g[:,2])
# 2.2.1 - Initialisation des conditions initiales
CI_old = 0.                   # Modifiable (par défaut 0.331 ou 0)
CI_new = 1.                   # Modifiable (par défaut 0.331 ou Nbr > 0) 

# 2.2.2 - Premier tir
start_time = time.time() # tic
g_old = RK4_Blasius(m, 0, etamax, CI_old)

# 2.2.3 - Boucle principale de la méthode du tir
cpt = 1      # Compteur d'itération
Erreur  = 1  # Condition d'enclenchement de la boucle

while Erreur > tol:       
    # Intégration du système
    g_new =  RK4_Blasius(m, 0., etamax, CI_new)

    # Interpolation linéaire : y = ax + b => a = Delta_y / Delta_x ; b = y - ax
    a = (g_new[-1,1] - g_old[-1,1])/(CI_new - CI_old) 
    b = g_new[-1,1] - a*CI_new
    
    # Nouvelles conditions initiales : x = (y - b)/a   
    CI_old = CI_new
    CI_new = (1 - b)/a

    # Itération suivante
    g_old = g_new
    Erreur = abs(g_new[-1,1] - 1)
        
    cpt += 1

DureeTotale = time.time() - start_time  # toc 

g = g_new

# 2.3 - Approximation de g_2 pour les petites valeurs de eta (0 < eta <~ 3)
i = 0
while eta_rk4[i] < 3:
    i = i+1
eta_court = eta_rk4[:i]

# 2.4 - Sorties
# 2.4.1 - Numériques
print('g_2(inf) : ', g[-1,1], 'à',tol,'près')          # facteur dimensionnel : U* 
print('g_3(0) : ', round(g[0,2],floor(-log(tol))))     # facteur dimensionnel : U* 
if build == 0 :
    print('Longueur (xmax) : ', xmax, 'm')
    print('Hauteur (ymax) : ', round(ymax,4), 'm')
print('Pas d\'intégration : ', round(etamax/(m-1),4), 'm')
print('Nombre d\'itérations : ', cpt) 
print('Durée de l\'intégration : ', round(DureeTotale,4), 's')
print('Durée moyenne d\'une itération : ', round(DureeTotale/(cpt),4), 's')

# 2.4.2 - Graphiques
# Voir code ci-dessous dans les différences finies
```

### 2.2 - Différences finies appliquées aux équations de Prandtl

#### 2.2.1 - Rappel des équations de Prandtl
Nous attaquons maintenant le système d'équations de Prandtl par la méthode des différences finies. Commençons par reprendre le système :

$$ \left\{ \;    
\begin{array}{l} 
\vec{\nabla} \cdot \vec{u} = 0   \\
u_x \frac{\partial u_x}{\partial x} + u_y \frac{\partial u_x}{\partial y} = \nu \frac{\partial^2 u_x}{\partial y^2}  \\
\end{array} 
\right.
$$

avec les conditions limites : 
<table style="border:none" width=100% >
    <tr style="border:none">
        <td style="border:none"><div align=right>Non glissement $ $</div></td>
        <td style="border:none">$ $ $u_x(x,0) = 0$</td>
    </tr>

    <tr style="border:none">
        <td style="border:none"><div align=right>Flux entrant uniforme $ $</div></td>
        <td style="border:none">$ $ $u_x(0,y) = U$ et $u_y(0,y) = 0$</td>
    </tr>

    <tr style="border:none">
        <td style="border:none"><div align=right>Imperméabilité de la plaque $ $</div></td>
        <td style="border:none">$ $ $u_y(x,0) = 0$</td>
    </tr>
    
    <tr style="border:none">
        <td style="border:none"><div align=right>Vitesse uniforme à l'infini $ $</div></td>
        <td style="border:none">$ $ $u_x(x,y \to \infty) \to U$</td>
    </tr>    
</table>

Rappelons que ce système n'est valaque qu'à grand nombre de Reynolds.

#### 2.2.2 - Schéma numérique utilisé

Nous utilisons le schéma proposé par <a href="#ref2" style="text-decoration:none">[2]</a>. Discrétisons le plan $(x,y)$ en construisant une grille dont le pas dans la direction $x$ est $h_x$ et de la même manière, $h_y$ sera le pas dans la direction $y$. On associe ensuite l'indice $i$ à l'abscisse et l'indice $j$ à l'ordonnée d'un nœud. Sur ces considérations, on peut écrire les différences centrées suivantes :

$$
\left( \frac{\partial u_x}{\partial y} \right)_{i,j} \approx \frac{u_x(i,j+1) - u_x(i,j-1)}{2 h_y} \\
\left( \frac{\partial^2 u_x}{\partial y^2} \right)_{i,j} \approx \frac{u_x(i,j+1) - 2u_x(i,j) + u_x(i,j-1)}{h_y^2}
$$

Où $u_x(i,j)$ est la valeur de $u_x$ au nœud $(i,j)$. La dérivée partielle dans la direction $x$ sera approchée par l'expression progressive suivante :

$$
\left( \frac{\partial u_x}{\partial x} \right)_{i,j} \approx \frac{u_x(i+1,j) - u_x(i,j)}{h_x} \\
$$

En substituant ces expressions dans le bilan d'impulsion on obtient la valeur de $u_x(i+1,j)$ comme suit :

$$
\fbox{$u_x(i+1,j) = u_x(i,j) + \nu \frac{u_x(i,j+1) - 2u_x(i,j) + u_x(i,j-1)}{u_x(i,j)} \frac{h_x}{h_y^2} - \frac{u_x(i,j+1) - u_x(i,j-1)}{2}\frac{u_y(i,j)}{u_x(i,j)}\frac{h_x}{h_y}$} 
$$

Connaissant $u_x(i+1,j)$ on obtient $u_y(i+1,j)$ via l'équation de continuité. Pour cela, on utilise d'une part la différence rétrograde en $y$ au nœud $(i+1,j)$ :

$$
\left( \frac{\partial u_y}{\partial y} \right)_{i+1,j} \approx \frac{u_y(i+1,j) - u_y(i+1,j-1)}{h_y}
$$

et d'autre part l'expression suivante pour la dérivée en $x$ :

$$
\left( \frac{\partial u_x}{\partial x} \right)_{i+1,j} \approx \frac{1}{2} \left[ \frac{u_x(i+1,j) - u_x(i,j)}{h_x} + \frac{u_x(i+1,j-1) - u_x(i,j-1)}{h_x}\right] \\
$$

On arrive alors à trouver $u_y(i+1,j)$ :

$$
\fbox{$u_y(i+1,j) = u_y(i+1,j-1) - \frac{1}{2} \left[ \frac{u_x(i+1,j) - u_x(i,j)}{h_x} + \frac{u_x(i+1,j-1) - u_x(i,j-1)}{h_x}\right] h_y$}
$$

Remarquons que pour la dérivée en $x$ on peut se restreindre à :

$$
\frac{u_x(i+1,j-1) - u_x(i,j-1)}{h_x}\\
$$

Ce dernier cas correspond à prendre schema = 1 plutôt que schema = 0 (par défaut) dans le code ci-dessous. Cependant, la solution obtenue sous cette restriction s'écarte légèrement du résultat obtenu par Runge-Kutta.

#### 2.2.3 - Critère de convergence
La convergence d'un schéma numérique est essentielle. On peut montrer <a href="#ref3" style="text-decoration:none">[3]</a> que les pas de discrétisation ne sont pas indépendants et doivent vérifier la relation suivante afin d'assurer la convergence du schéma ci-dessus :

$$
h_x \leq \frac{h_y^2}{2\nu}u_x(i,j)
$$

#### 2.2.4 - Intégration du système de Prandtl

Attention, avant la première éxécution de ce code, il faudra avoir fait tourner Runge-Kutta au moins une fois pour définir certaines variables.

```python
# 2.4.2 - Graphiques
figDF, axesDF = plt.subplots(2,sharex = True, figsize=(17,10))  

u_rk4 = g[:,1]
v_rk4 = sqrt((nu*U)/(xmax/2))*0.5*(eta_rk4[:]*g[:,1] - g[:,0])/U

# u : facteur dimensionnel U*
axesDF[0].plot(eta_rk4, u_rk4,'b') #  
axesDF[0].set_ylabel('$u/U_\infty$')
axesDF[0].set_title('Vitesse longitudinale et comportement près de la plaque')
axesDF[0].xaxis.grid(True,'major')
axesDF[0].yaxis.grid(True,'major')

# v : facteur dimensionnel sqrt((nu*U)/(xmax/2))* /!\ dans le build 1 à la consistance avec le build 0
axesDF[1].plot(eta_rk4,v_rk4,'b',label="Par RK4")
axesDF[1].set_ylabel('$v/U_\infty$')
axesDF[1].set_title('Vitesse transversale')
axesDF[1].xaxis.grid(True,'major')
axesDF[1].yaxis.grid(True,'major')

# Comportement à l'origine : facteur dimensionnel U*
axesDF[0].plot(eta_court, eta_court*g[0,2],'r')

################################################################

# --------------------------------------------------------------
# 3 - Résolution des équations de Prandtl par différences finies
# --------------------------------------------------------------

# Plan
# ----

# 3.1 - Initialisation
# 3.1.1 - Paramètrisation de la méthode des différences finies 
# 3.1.2 - Paramètres physiques

# 3.2 - Résolution numérique
    # 3.2.1 - Initialisation 
    # 3.2.2 - Condition initiale non nulle
    # 3.2.3 - Différences finies : _i+1,j+1 = [j-1,i+1]
    
# 3.3 - Sorties
    # 3.3.1 - Numériques
    # 3.3.2 - Graphique

# --------------------------------------------------------------  

# 3.1 - Initialisation     
# 3.1.1 - Paramètrisation de la méthode des différences finies 
hx = 10**(-4)   # Modifiable (par défaut 10**(-4))
hy = 10**(-4)   # Modifiable (par défaut 10**(-4))

m = 7000        # Modifiable (par défaut 7000)
n = 60          # Modifiable (par défaut 60)

schema = 0      # Modifiable : 0 ou 1 (par défaut 0)

xmax = round(m*hx,8)
ymax = round(n*hy,8)

# 3.1.2 - Paramètres physiques
U = 10          # Modifiable (par défaut 10 m/s)
nu = 10**(-5)   # Modifiable (par défaut 10**(-5) m²/s)

# 3.2 - Résolution numérique
# 3.2.1 - Initialisation 
u = np.zeros((n, m))
v = np.zeros((n, m))

# 3.2.2 - Condition initiale non nulle
u[0:-1,0] = U # bord gauche
u[0,0:] = U   # bord haut

# 3.2.3 - Différences finies : _i+1,j+1 = [j-1,i+1]
if schema == 0 :
    start_time = time.time()
    for i in range(0,m-1) : # Colonne <--> axe x

        # u_i+1,j
        dy2u = (u[n-1:1:-1,i] -2*u[n-2:0:-1,i] + u[n-3::-1,i])/hy**2
        dyu = (u[n-3::-1,i] - u[n-1:1:-1,i])/(2*hy)

        u[n-2:0:-1,i+1] = u[n-2:0:-1,i] + hx*(nu*dy2u[:] - dyu[:]*v[n-2:0:-1,i])/u[n-2:0:-1,i]

        # v_i+1,j
        for j in range(1,n-1) : # Ligne <--> axe y
            # Axe y pointe vers le haut et non vers le bas
            j = n-1 - j # pour aller de n-2 à 1 

            dxu_a = (u[j,i+1] - u[j,i])/hx 
            dxu_b = (u[j+1,i+1] - u[j+1,i])/hx 

            v[j,i+1] = v[j+1,i+1] - 0.5*(dxu_a + dxu_b)*hy
            
elif schema == 1 :
    start_time = time.time()
    for i in range(0,m-1) : # Colonne <--> axe x

        # u_i+1,j
        dy2u = (u[n-1:1:-1,i] -2*u[n-2:0:-1,i] + u[n-3::-1,i])/hy**2
        dyu = (u[n-3::-1,i] - u[n-1:1:-1,i])/(2*hy)

        u[n-2:0:-1,i+1] = u[n-2:0:-1,i] + hx*(nu*dy2u[:] - dyu[:]*v[n-2:0:-1,i])/u[n-2:0:-1,i]

        # v_i+1,j
        for j in range(1,n-1) : # Ligne <--> axe y
            # Axe y pointe vers le haut et non vers le bas
            j = n-1 - j # pour aller de n-2 à 1 

            dxu = (u[j+1,i+1] - u[j+1,i])/hx 

            v[j,i+1] = v[j+1,i+1] - dxu*hy

DureeTotale = time.time() - start_time                                         

# 3.3 - Sorties
# 3.3.1 - Numériques
print('Longueur (xmax) : ', xmax, 'm')
print('Pas sur x : ', hx, 'm')
print('Hauteur (ymax) : ', round(ymax,4), 'm')
print('Pas sur y : ', hy, 'm')
print('Pas sur eta : ', round(sqrt(U/nu)*hy/sqrt(xmax/2),4), 'm',)
print('Durée de l\'intégration : ', round(DureeTotale,4), 's')

# 3.3.2 - Graphique
y = np.linspace(0, ymax, n+1) # +1 pour avoir le même nombre de point qu'avec RK4
eta = sqrt(U/nu)*(ymax - y[1:])/sqrt(xmax/2) # eta(x = xmax/2, y), à l'envers pour que le graphe soit à l'endroit

u_df = u[:,round(m/2)]/U
v_df = v[:,round(m/2)]/U

# u    
axesDF[0].plot(eta[1:],u_df[1:],'g*')
axesDF[0].set_ylabel('$u/U_\infty$')

# v
axesDF[1].plot(eta[1:],v_df[1:],'g*',label="Par D.F.")
axesDF[1].set_xlabel('$\eta$')
axesDF[1].set_ylabel('$v/U_\infty$')

plt.legend(bbox_to_anchor=(0., 1., 1., .1), loc=3,
           ncol=2, mode="expand", borderaxespad=1.5)

# Ecart maximal entre les courbes
ecart_u = abs(u_rk4[:-1] - u_df[::-1])  
ecart_v = abs(v_rk4[:-1] - v_df[::-1]) 
    
ecart_rel_u = 1 - abs(u_rk4[1:-2] - ecart_u[1:-1])/u_rk4[1:-2]
ecart_rel_v = 1 - abs(v_rk4[1:-2] - ecart_v[1:-1])/v_rk4[1:-2]

print('')
print('Ecart maximal entre RK4 et différences finies')
print('---------------------------------------------')
print('Sur la vitesse longitudinale : ', max(ecart_u), '-->', max(100*ecart_rel_u), '%') 
print('Sur la vitesse transervale : ', max(ecart_v), '-->', max(100*ecart_rel_v), '%')
```

## Conclusions

Chacune des méthodes présentée ici a un "handicap", c'est à dire que RK4 doit être couplée avec le tir car il lui manque une condition initiale et que les différences finies travaillent en deux dimensions, ce qui les empêche d'exploiter complètement la puissance des boucles implicites sous peine de fournir de mauvais résultats pour la vitesse longitudinale.

Nous obtenons des graphes "identiques" via les deux méthodes de résolution du problème lorsque les options par défaut sont choisies. Pour un même pas d'intégration sur la variable $\eta$ et donc pour une même précision, RK4 l'emporte de près de deux ordres de grandeur en vitesse de calcul : quelques centièmes de secondes contre quelques secondes pour les différences finies.

Outre cela, trouver un schéma de différences finies *convergent* s'est avéré ardu. En effet, dériver un tel schéma est aisé en discrétisant via la méthode d'Euler, mais ceci n'est guère satisfaisant dans la mesure où cette dernière est inconditionnellement instable. Il faut donc discrétiser les dérivées par des formules de différences centrées qui impactent directement la complexité de la dérivation algébrique des schémas. Quant à RK4, elle ne fait intervenir aucune difficulté particulière : le schéma est parfaitement connu et ne s'applique qu'à une seule équation différentielle ordinnaire du $3^{ème}$ ordre.

Les deux méthodes sont finalement très efficaces. Néanmoins, si les équations aux dérivées partielles à résoudre peuvent se ramener à des équations ordinnaires, il est évident que ces dernières seront toujours plus simples et plus rapides à traiter.
